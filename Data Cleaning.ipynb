```python
import pandas as pd
import numpy as np
jobs = pd.read_csv('finaldf.csv', usecols = ['Title','Mark','User','Keywords'])
jobs.sample(5)
```

```python
jobs.isnull().sum()
```

```python
jobs['Keywords'] = jobs['Keywords'].str.replace("#", "")
jobs['Keywords'] = jobs['Keywords'].str.replace("\n\n", " ")
jobs['Keywords'] = jobs['Keywords'].str.replace("\n", " ")
jobs.sample(5)
```

```python
jobs['User'] = jobs['User'].replace(np.nan, '')
jobs['Keywords'] = jobs['Keywords'].replace(np.nan, '')
jobs.sample(5)
```

```python
jobs["Keywordlist"] = jobs["Keywords"].str.split(" ")
jobs.sample(5)
```

```python
# jobs.drop("Keywords", axis=1, inplace=True)
```

```python
jobs.sample(5)
```

```python
jobs["Userlist"] = jobs["User"].str.split(" ")
jobs.sample(5)
```

```python
# jobs.drop("User", axis=1, inplace=True)
```

```python
jobs["desclist"] = jobs["Mark"].str.split(" ")
jobs.sample(5)
```

```python
jobs.isnull().sum()
```

```python
# jobs.drop("Mark", axis=1, inplace=True)
```

```python
jobs['tags'] = jobs['Userlist'] + jobs['Keywordlist'] + jobs['desclist']
jobs.sample(5)
```

```python
jobs.sample(5)
```

```python
new = jobs.drop(columns=['User','Mark','Keywords','Keywordlist','Userlist','desclist'])
```

```python
new.sample(5)
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
new.sample(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
new[new['Title'] == '{PS4} [SQTR] Sky Limits #2 (Squall-TR)'].index[0]
```

```python
def recommend(jobs):
    index = new[new['Title'] == jobs].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].Title)
```

```python
recommend('{P$4} AirGermany Parkour #68 - SULTAN 4x4 [DaveStylemh]')
```
